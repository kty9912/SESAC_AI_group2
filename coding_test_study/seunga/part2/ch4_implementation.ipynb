```python
# ì…ë ¥
n = int(input())
moves = str(input().split())

now = [1,1]

# ì´ë™
for move in moves: # moves í•˜ë‚˜ì”© í™•ì¸
    if move == 'L':
        if now[1] > 1:
            now[1] -= 1
    elif move == 'R':
        if now[1] < n:
            now[1] += 1
    elif move == 'U': # 'up'ì€ ë°‘ìœ¼ë¡œ ê°€ì•¼ ì»¤ì§
        if now[0] > 1:
            now[0] -= 1
    elif move == 'D':
        if now[0] < n:
            now[0] += 1

# ì¶œë ¥
print(now[0], now[1])
```

```python
# ë‚˜ì´íŠ¸ì˜ í˜„ì¬ ìœ„ì¹˜ ì…ë ¥ ë°›ê¸°
knight = input()
row = int(knight[1])
column = int(ord(knight[0])) - int(ord('a')) + 1 ##

count = 0

# ë‚˜ì´íŠ¸ì˜ ì´ë™ ê°€ëŠ¥ ê²½ìš°ì˜ ìˆ˜ ì„¸ê¸°
# ë‚˜ì´íŠ¸ì˜ ì´ ì´ë™ ê°€ëŠ¥ ê²½ìš°ì˜ ìˆ˜: 8ê°€ì§€
moves = [(-1, 2), (1, 2), (-1, -2), (1, -2), (-2, -1), (-2, 1), (2, -1), (2, 1)]

# í˜„ì¬ ìœ„ì¹˜ì—ì„œ ê°€ëŠ¥í•œ ê²½ìš°ë¥¼ í™•ì¸í•´ì„œ ì¹´ìš´íŠ¸ì— ë„£ê¸°
for move in moves:
    next_row = row + move[0]
    next_column = column + move[1]
    
    if 1 <= next_row <= 8 and 1<= next_column <= 8:
        count += 1

# ì¶œë ¥
print(count)
```

###### - ë‘˜ì§¸ ì¤„ì— ê²Œì„ ìºë¦­í„°ê°€ ìˆëŠ” ì¹¸ì˜ ì¢Œí‘œ (A, B)ì™€ ë°”ë¼ë³´ëŠ” ë°©í–¥ dê°€ ê°ê° ì„œë¡œ ê³µë°±ìœ¼ë¡œ êµ¬ë¶„í•˜ì—¬ ì£¼ì–´ì§„ë‹¤. ë°©í–¥ dì˜ ê°’ìœ¼ë¡œëŠ” ë‹¤ìŒê³¼ ê°™ì´ 4ê°€ì§€ê°€ ì¡´ì¬í•œë‹¤.
###### 0: ë¶, 1: ë™, 2: ë‚¨, 3: ì„œìª½
###### - ì…‹ì§¸ ì¤„ë¶€í„° ë§µì´ ìœ¡ì§€ì¸ì§€ ë°”ë‹¤ì¸ì§€ì— ëŒ€í•œ ì •ë³´ê°€ ì£¼ì–´ì§„ë‹¤. Nê°œì˜ ì¤„ì— ë§µì˜ ìƒíƒœê°€ ë¶ìª½ë¶€í„° ë‚¨ìª½ ìˆœì„œëŒ€ë¡œ, ê° ì¤„ì˜ ë°ì´í„°ëŠ” ì„œìª½ë¶€í„° ë™ìª½ ìˆœì„œëŒ€ë¡œ ì£¼ì–´ì§„ë‹¤. ë§µì˜ ì™¸ê³½ì€ í•­ìƒ ë°”ë‹¤ë¡œ ë˜ì–´ ìˆë‹¤. 0: ìœ¡ì§€, 1: ë°”ë‹¤
###### - ì²˜ìŒì— ê²Œì„ ìºë¦­í„°ê°€ ìœ„ì¹˜í•œ ì¹¸ì˜ ìƒíƒœëŠ” í•­ìƒ ìœ¡ì§€ì´ë‹¤.

###### ì¶œë ¥ ì¡°ê±´
###### - ì²«ì§¸ ì¤„ì— ì´ë™ì„ ë§ˆì¹œ í›„ ìºë¦­í„°ê°€ ë°©ë¬¸í•œ ì¹¸ì˜ ìˆ˜ë¥¼ ì¶œë ¥í•œë‹¤.

###### 1 1 1 1  # ì²« ì¤„ì€ ëª¨ë‘ ë°”ë‹¤

###### 1 1 1 1  # ë„·ì§¸ ì¤„ì€ ëª¨ë‘ ë°”ë‹¤

```python
# N, Mì„ ê³µë°±ìœ¼ë¡œ êµ¬ë¶„í•˜ì—¬ ì…ë ¥ë°›ê¸°
n, m = map(int, input().split())

# ë°©ë¬¸í•œ ìœ„ì¹˜ë¥¼ ì €ì¥í•˜ê¸° ìœ„í•œ ë§µì„ ìƒì„±í•˜ë©° 0ìœ¼ë¡œ ì´ˆê¸°í™”
d = [[0] * m for _ in range(n)]

# í˜„ì¬ ìºë¦­í„°ì˜ X ì¢Œí‘œ, Y ì¢Œí‘œ, ë°©í–¥ì„ ì…ë ¥ë°›ê¸°
x, y, direction = map(int, input().split())

# í˜„ì¬ ì¢Œí‘œ ë°©ë¬¸ ì²˜ë¦¬
d[x][y] = 1

# ì „ì²´ ë§µ ì •ë³´ë¥¼ ì…ë ¥ë°›ê¸°
array = []
for i in range(n):
    array.append(list(map(int, input().split())))

# ë¶, ë™, ë‚¨, ì„œ ë°©í–¥ ì •ì˜
dx = [-1, 0, 1 ,0]
dy = [0, 1 ,0, -1]

# ì™¼ìª½ìœ¼ë¡œ íšŒì „
def turn_left():
    global direction
    direction -= 1
    if direction == -1:
        direction = 3

# ì‹œë®¬ë ˆì´ì…˜ ì‹œì‘
count = 1
turn_time = 0
while True:
    
    # ì™¼ìª½ìœ¼ë¡œ íšŒì „
    turn_left()
    nx = x + dx[direction]
    ny = y + dy[direction]

    # íšŒì „í•œ ì´í›„ ì •ë©´ì— ê°€ë³´ì§€ ì•Šì€ ì¹¸ì´ ì¡´ì¬í•˜ëŠ” ê²½ìš° ì´ë™
    if d[nx][ny] == 0 and array[nx][ny] == 0:
        d[nx][ny] = 1
        x = nx
        y = ny
        count += 1
        turn_time = 0
        continue

    #íšŒì „í•œ ì´í›„ ì •ë©´ì— ê°€ë³´ì§€ ì•Šì€ ì¹¸ì´ ì—†ê±°ë‚˜ ë°”ë‹¤ì¸ ê²½ìš°
    else:
        turn_time += 1
    
    # ë„¤ ë°©í–¥ ëª¨ë‘ ê°ˆ ìˆ˜ ì—†ëŠ” ê²½ìš°
    if turn_time == 4:
        nx = x - dx[direction]
        ny = y - dy[direction]

        # ë’¤ë¡œ ê°ˆ ìˆ˜ ìˆë‹¤ë©´ ì´ë™í•˜ê¸°
        if array[nx][ny] == 0:
            x = nx
            y = ny 

        # ë’¤ê°€ ë°”ë‹¤ë¡œ ë§‰í˜€ìˆëŠ” ê²½ìš°
        else:
            break
        turn_time = 0

# ì •ë‹µ ì¶œë ¥
print(count)
```
